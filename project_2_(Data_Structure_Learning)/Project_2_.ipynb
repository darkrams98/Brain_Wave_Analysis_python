```python
# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████
# import json class and read JSON file :
# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████

import json

with open("samples_input.json", "r") as json_file:
    data = json.load(json_file)

# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████
# Iterate through the JSON data and print : 
# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████
# Iterate through the JSON data and print theme

for entry in data:
    name = entry["name"]
    birth_year = entry["birth_year"]
    height = entry["height"]
    weight = entry["weight"]

    # Do something with the data, for example, print it
    print(f"Name: {name}, Birth Year: {birth_year}, Height: {height}, Weight: {weight}")
    
# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████
# Convert to pandas Data frame and show head :
# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████

import pandas as pd

# Convert the JSON data to a Pandas DataFrame
df = pd.DataFrame(data)

# Now, you have your data in a Pandas DataFrame, and you can work with it as you like.
df.head()
```

```python
# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████
# read old data from json file : 
# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████

with open("samples_old_inputs.json", "r") as json_file:
    data_old = json.load(json_file)
    
# Convert the JSON data to a Pandas DataFrame
df_old = pd.DataFrame(data_old)

# Now, you have your data in a Pandas DataFrame, and you can work with it as you like.
df_old.head()

# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████
# concat column names to attach the main dataset
# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████


import numpy as np

# get columns : 
new_cols = df.columns
old_cols = df_old.columns

new_cols_np = np.array(new_cols)
old_cols_np = np.array(old_cols)

cols_concat = np.concatenate([new_cols_np,old_cols_np ], axis=0)

cols_concat

# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████
# concat data frame and create main dataset :
# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████

# concat :
df_concat_result = pd.concat([df, df_old] ,ignore_index=True, axis=1)
df_concat_result.columns = cols_concat
df_concat_result.head(10)
```

```python
# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████
# calculate age by birthday
# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████
# import date class
from datetime import datetime

# تابع برای محاسبه سن افراد
def calculate_age(birth_year):
    birth_date = datetime.strptime(birth_year, "%d/%m/%y")
    current_date = datetime.now()
    age = current_date.year - birth_date.year
    if (current_date.month, current_date.day) < (birth_date.month, birth_date.day):
        age -= 1
    return age

# create year column :
df_concat_result["age_year"] = df_concat_result["birth_year"].apply(calculate_age)

# create day column : 
df_concat_result["age_day"] = (datetime.now().day - df_concat_result["birth_year"].str[:2].astype(int)).abs().astype(str)

# نمایش داده‌های نهایی
print(df_concat_result[["name", "age_year", "age_day"]])
```

```python
# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████
# set password for login : 
# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████
from cryptography.fernet import Fernet

# تولید یک کلید رمزگذاری جدید
key = Fernet.generate_key()
fernet = Fernet(key)

# رمزنگاری رمز عبور
password = "Darkrams"
encrypted_password = fernet.encrypt(password.encode())

# ذخیره کردن رمز عبور رمزنگاری شده در یک فایل JSON
data = {"password": encrypted_password.decode("utf-8")}
with open("password.json", "w") as json_file:
    json.dump(data, json_file)

# open password
with open("password.json", "r") as json_file:
    data = json.load(json_file)
    encrypted_password = data["password"]

# رمزگشایی رمز عبور
decrypted_password = fernet.decrypt(encrypted_password.encode()).decode("utf-8")
# print password
print("Password : ", decrypted_password)
```

```python

# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████
# password validation :
    # password for test : Darkrams
# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████

# Open the password JSON file and load the data
with open('password.json') as f:
  data = json.load(f) 

# Get the encrypted password from the data
encrypted_password = data['password']  

# Initialize attempts counter for password tries
attempts = 0

# Loop to allow up to 4 password attempts
while attempts < 4:

  # Get password input from user
  password = input("Please enter your password: ")

  try:
    # Decrypt the stored encrypted password
    decrypted_password = fernet.decrypt(encrypted_password.encode()).decode()

    # Check if input password matches decrypted password
    if password == decrypted_password:

      # Get user's name input
      name = input("please input your name :")
      
      # Open user data JSON file and load  
      with open('output_data.json') as f:
        main_data = json.load(f)

      # Convert loaded data to Pandas DataFrame
      df = pd.DataFrame(main_data, columns=['name', 'birth_year', 'height', 'weight', 'num_children',  
                        'blood_pressure', 'hospitalized', 'genetic_disease', 'fbs',
                        'age_year', 'age_day'])
      
      # Check if name exists in loaded data
      if name in df['name'].values:
      
        # Print success message
        print(f"Welcome {name}! Your information is successfully stored.")
        
        # Break out of loop on correct password
        break 

      else:
        # Name not found error message
        print("Your name was not previously entered.")
        break

    else:
      # Print invalid password error
      print("Incorrect password. Please try again.")

  except:
    # Print invalid password for any errors
    print("Incorrect password. Please try again.")

  # Increment password attempts 
  attempts += 1

# If 4 failed attempts, print lock out message 
if attempts == 4:
  print("Too many failed attempts. System locked.")
```
