```python
# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████
# calculate mean and sd 
# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████

eeg_np_trans = eeg_np.transpose()
n_trials = len(eeg_np_trans)
mn = eeg_np_trans.mean(0)                               # Compute the mean signal across trials (the ERP).
sd = eeg_np_trans.std(0)                                # Compute the std of the signal across trials.
sdmn = sd / np.sqrt(n_trials)                       # Compute the std of the mean.

print(n_trials)
print(len(mn))
print(len(sd))
print(len(sdmn))
```

```python
# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████
# calculate direction of test : 
# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████

# sampling frequency : 
fcy = 500

# direction in second and minute 
time_sec = n_trials / fcy
time_min = time_sec / 60

# مقیاس زمان به ثانیه برای هر نمونه
time_scale = 1 / fcy

# ایجاد ستون "Time" برای داده‌ها به وسیله مقیاس زمان
time_column = np.arange(0, n_trials) * time_scale

print("Test direction : " , int(time_sec) , " second ", " OR ", int(time_min) , " minute")
time_column = time_column.round(3)
time_column
```

```python
def extract_direction( data , time_start , time_stop , sampling_frequency) :
    # ██████████████████████████████████████████████████████████████████████████████████████████████████████████████
    # calculate time_start and time stop indexes :
    # ██████████████████████████████████████████████████████████████████████████████████████████████████████████████

    sample_index_start = int(time_start * sampling_frequency)
    sample_index_stop = int(time_stop * sampling_frequency)

    # ██████████████████████████████████████████████████████████████████████████████████████████████████████████████
    # create subset with time_start and time_stop indexes :
    # ██████████████████████████████████████████████████████████████████████████████████████████████████████████████

    eeg_df_subset = eeg_np[:, sample_index_start : sample_index_stop + 1]
    eeg_time_between = time_column[sample_index_start : sample_index_stop + 1]

    # ██████████████████████████████████████████████████████████████████████████████████████████████████████████████
    # get output for validate data structure :
    # ██████████████████████████████████████████████████████████████████████████████████████████████████████████████

    # print(eeg_df_subset)
    # print(eeg_time_between)
    print("start input (second) : ", eeg_time_between[0])
    print("end input (second) : ", eeg_time_between[-1])
    print("start input index (int) : ", list(time_column).index(eeg_time_between[0]))
    print("end input index (int) : ", list(time_column).index(eeg_time_between[-1]))
    print(np.shape(eeg_df_subset))
    print(np.shape(eeg_time_between))
    
    # return : 
    return {"start_index" : sample_index_start, "stop_index" : sample_index_stop, "data" : eeg_df_subset, "time_dir": eeg_time_between}
```

```python
# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████
# extract direction :  
# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████

range_1 = extract_direction(eeg_np, 30, 40, fcy)
range_2 = extract_direction(eeg_np, 150, 160 ,fcy)
range_3 = extract_direction(eeg_np, 295, 300, fcy)
```

```python
# print output keys : 
print(range_1.keys())
print(range_2.keys())
print(range_3.keys())
```

```python
# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████
# data to numpy array : 
# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████

# convert to numpy array 
range_1_np = np.array(range_1["data"])
range_2_np = np.array(range_2["data"])
range_3_np = np.array(range_3["data"])

# print shapes
print(np.shape(range_1_np))
print(np.shape(range_2_np))
print(np.shape(range_3_np))


# remove mean : 

for i in range(0,29,1) :
    range_1_np[i] = range_1_np[i] - mn[i]
    range_2_np[i] = range_2_np[i] - mn[i]
    range_3_np[i] = range_3_np[i] - mn[i]
```

```python
# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████
# convert to data frame : 
# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████

# convert to dataframe
channel_labels = ['Fp1', 'Fp2', 'F7', 'F3', 'Fz', 'F4', 'F8', 'C3', 'Cz', 'Cp6', 'Cp2', 'Oz','Cp5', 'Cp1', 'T8', 'Fcz', 'T7', 'Fc6', 'Fc2', 'Fc1', 'Fc5', 'O2', 'O1','P8', 'P4', 'Pz', 'P3', 'P7', 'C4']
range_1_df = pd.DataFrame(range_1_np, columns=range_1["time_dir"], index=channel_labels)
range_2_df = pd.DataFrame(range_2_np, columns=range_2["time_dir"], index=channel_labels)
range_3_df = pd.DataFrame(range_3_np, columns=range_3["time_dir"], index=channel_labels)

# print shapes 
print(range_1_df.shape)
print(range_2_df.shape)
print(range_3_df.shape)
```

```python
# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████
# create subplot : 
# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████

# set defaults 
Vertical_space = 5


# function to plot eeg signals
def plot_subplot(EEG_signal, time_axis) :
    for channel in range(EEG_signal.shape[0]):
      channel_signal = EEG_signal.iloc[channel, :] + channel * Vertical_space
      plt.plot(time_axis, channel_signal)
      channel_labels = ['Fp1', 'Fp2', 'F7', 'F3', 'Fz', 'F4', 'F8', 'C3', 'Cz', 'Cp6', 'Cp2', 'Oz','Cp5', 'Cp1', 'T8', 'Fcz', 'T7', 'Fc6', 'Fc2', 'Fc1', 'Fc5', 'O2', 'O1','P8', 'P4', 'Pz', 'P3', 'P7', 'C4']
      channel_positions = np.arange(EEG_signal.shape[0]) * Vertical_space
      plt.yticks(channel_positions, channel_labels)
      plt.xlabel('Time(s)')
      plt.ylabel('Channel')
      plt.title(f'EEG signal range {time_axis[0]} to {time_axis[-1]}')
    plt.show()
```

```python
# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████
# create subplot : 
# ██████████████████████████████████████████████████████████████████████████████████████████████████████████████
plt.figure(figsize=(12 , 12))
plot_subplot(range_1_df, range_1["time_dir"])
plt.figure(figsize=(12 , 12))
plot_subplot(range_2_df, range_2["time_dir"])
plt.figure(figsize=(12 , 12))
plot_subplot(range_3_df, range_3["time_dir"])
```
