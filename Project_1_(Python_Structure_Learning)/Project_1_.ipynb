```python
# ███████████████████████████████████████████████████████
# ███████████████████████████████████████████████████████
# import json class and read JSON file :
# ███████████████████████████████████████████████████████
# ███████████████████████████████████████████████████████

import json

with open("Samples_Get_Input.json", "r") as json_file:
    data = json.load(json_file)
```

```python
# ███████████████████████████████████████████████████████
# ███████████████████████████████████████████████████████
# Iterate through the JSON data and print : 
# ███████████████████████████████████████████████████████
# ███████████████████████████████████████████████████████
# Iterate through the JSON data
for entry in data:
    name = entry["name"]
    birth_year = entry["birth_year"]
    height = entry["height"]
    weight = entry["weight"]

    # Do something with the data, for example, print it
    print(f"Name: {name}, Birth Year: {birth_year}, Height: {height}, Weight: {weight}")
```

```python
# ███████████████████████████████████████████████████████
# ███████████████████████████████████████████████████████
# Convert to pandas Data frame and show head :
# ███████████████████████████████████████████████████████
# ███████████████████████████████████████████████████████

import pandas as pd

# Convert the JSON data to a Pandas DataFrame
df = pd.DataFrame(data)

# Now, you have your data in a Pandas DataFrame, and you can work with it as you like.
df.head()
```

```python
# ███████████████████████████████████████████████████████
# ███████████████████████████████████████████████████████
# read old data from json file : 
# ███████████████████████████████████████████████████████
# ███████████████████████████████████████████████████████

with open("Samples_Old_Data.json", "r") as json_file:
    data_old = json.load(json_file)
    
# Convert the JSON data to a Pandas DataFrame
df_old = pd.DataFrame(data_old)

# Now, you have your data in a Pandas DataFrame, and you can work with it as you like.
df_old.head()
```

```python
# ███████████████████████████████████████████████████████
# ███████████████████████████████████████████████████████
# concat columns name to attach the main dataset
# ███████████████████████████████████████████████████████
# ███████████████████████████████████████████████████████


import numpy as np

# get columns : 
new_cols = df.columns
old_cols = df_old.columns

new_cols_np = np.array(new_cols)
old_cols_np = np.array(old_cols)

cols_concat = np.concatenate([new_cols_np,old_cols_np ], axis=0)

cols_concat
```

```python
# ███████████████████████████████████████████████████████
# ███████████████████████████████████████████████████████
# concat data frame :
# ███████████████████████████████████████████████████████
# ███████████████████████████████████████████████████████

# concat :
df_concat_result = pd.concat([df, df_old] ,ignore_index=True, axis=1)
df_concat_result.columns = cols_concat
df_concat_result.head()

```

```python
# ███████████████████████████████████████████████████████
# ███████████████████████████████████████████████████████
# calculate age with birth year :
# ███████████████████████████████████████████████████████
# ███████████████████████████████████████████████████████

from datetime import date

def calculate_age(birth_years):
  today = date.today() 
  ages = []

  for birth_year in birth_years:
    age = today.year - birth_year
    ages.append(age)

  return ages


ages = calculate_age(df_concat_result["birth_year"])

df_concat_result["age"] = ages
df_concat_result.head()
```

<h1 style="margin:20px 0px 20px 0px; width:100%; text-align:center; color:white; height:80px; font-size:45px; font-weight:bold; test-justify:center; background-color:blue;border-radius:2px ;padding-top:15px;">Task 5</h1>

```python
# ███████████████████████████████████████████████████████
# ███████████████████████████████████████████████████████
# calculate BMI with weight and height : 
# ███████████████████████████████████████████████████████
# ███████████████████████████████████████████████████████

def calculate_bmi(height_list, weight_list):
    # بررسی تعداد عناصر داخل هر لیست
    if len(height_list) != len(weight_list):
        return "تعداد عناصر در دو لیست باید یکسان باشد."

    # ساختن یک لیست برای ذخیره BMI هر فرد
    bmi_list = []

    # محاسبه BMI برای هر فرد و اضافه کردن آن به لیست BMI
    for i in range(len(height_list)):
        height = height_list[i] / 100  # تبدیل قد به متر
        weight = weight_list[i]
        bmi = weight / (height ** 2)
        bmi_list.append(bmi)

    return bmi_list



# فراخوانی تابع و نمایش نتایج
bmi_df_concat_results = calculate_bmi(df_concat_result["height"], df_concat_result["weight"])
list_bmi = []
for i, bmi in enumerate(bmi_df_concat_results):
    
    list_bmi.append(bmi.round(2))

print(list_bmi)

```

```python
# ███████████████████████████████████████████████████████
# ███████████████████████████████████████████████████████
# concat BMI and main data : 
# ███████████████████████████████████████████████████████
# ███████████████████████████████████████████████████████

# reshape bmi into 2d array then convert to data frame :
list_bmi = np.array(list_bmi)
list_bmi.reshape((1, len(list_bmi)))
list_bmi = pd.DataFrame(list_bmi)
list_bmi.head()



# concat BMI and main data :
df_concat_result = pd.concat([df_concat_result, list_bmi] ,ignore_index=True, axis=1)
# add new column to columns list : 
cols_concat = np.append(cols_concat, values=['age','bmi'])
df_concat_result.columns = cols_concat
df_concat_result.head()
```
